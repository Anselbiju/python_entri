```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/Users/anselbiju/Downloads/myexcel.xlsx"
df = pd.read_excel(file_path, sheet_name="myexcel.csv")

df.info()
df.head()
```

```python

np.random.seed(42)

df["Height"] = np.random.randint(150, 181, size=df.shape[0])

df.head()
```

```python
team_distribution = df["Team"].value_counts()
team_percentage = (team_distribution / len(df)) * 100

team_df = pd.DataFrame({"Employee Count": team_distribution,"Percentage": team_percentage.round(2)})
team_df
```

```python
# Bar Chart
plt.figure(figsize=(14,9))
sns.barplot(x=team_df.index, y="Employee Count", data=team_df)
plt.title("Employee Distribution by Team")
plt.ylabel("Number of Employees")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8,10))
plt.pie(team_df["Employee Count"], labels=team_df.index, autopct='%1.1f%%', startangle=170)
plt.title("Employee Percentage Distribution by Team")
plt.axis('equal')
plt.show()
```

```python
#count employees per position
position_distribution = df["Position"].value_counts()
position_distribution
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=position_distribution.index, y=position_distribution.values)
plt.title("Employee Distribution by Position")
plt.ylabel("Number of Employees")
plt.xlabel("Position")
plt.xticks(rotation=45)
plt.show()
```

```python
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 100]
age_labels = ["<20", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80+"]

df["Age Group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=False)

age_group_distribution = df["Age Group"].value_counts().sort_index()
age_group_distribution
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values)
plt.title("Employee Distribution by Age Group")
plt.ylabel("Number of Employees")
plt.xlabel("Age Group")
plt.show()
```

```python
# Fill missing salary values with 0
df['Salary'] = df['Salary'].fillna(0)
```

```python
salary_by_team = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
salary_by_team
```

```python
salary_by_position = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
salary_by_position
```

```python

# Bar chart for salary by position
plt.figure(figsize=(10,6))
sns.barplot(x=salary_by_position.index, y=salary_by_position.values)
plt.title("Total Salary Expenditure by Position")
plt.ylabel("Total Salary")
plt.xlabel("Position")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for salary by team
plt.figure(figsize=(14,11))
sns.barplot(x=salary_by_team.index, y=salary_by_team.values)
plt.title("Total Salary Expenditure by Team")
plt.ylabel("Total Salary")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.show()
```

```python
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
